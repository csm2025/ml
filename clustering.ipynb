```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.metrics import accuracy_score

# Load the Iris dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target

# Helper function to print predictions and evaluate accuracy
def evaluate_clustering(predictions, true_labels):
    accuracy = accuracy_score(true_labels, predictions)
    print(f"Accuracy: {accuracy}")
    print("Predictions:")
    print(predictions)
    print("True Labels:")
    print(true_labels)
    print("\n")

# k-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans_predictions = kmeans.fit_predict(X)
print("k-Means Clustering:")
evaluate_clustering(kmeans_predictions, y)

# Agglomerative clustering
agglomerative = AgglomerativeClustering(n_clusters=3)
agg_predictions = agglomerative.fit_predict(X)
print("Agglomerative Clustering:")
evaluate_clustering(agg_predictions, y)

# DBSCAN clustering
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan_predictions = dbscan.fit_predict(X)
print("DBSCAN Clustering:")
evaluate_clustering(dbscan_predictions, y)

# Plotting the results
plt.figure(figsize=(15, 5))

# Plot k-Means clustering
plt.subplot(131)
plt.scatter(X[:, 0], X[:, 1], c=kmeans_predictions, cmap='viridis')
plt.title("k-Means Clustering")

# Plot Agglomerative clustering
plt.subplot(132)
plt.scatter(X[:, 0], X[:, 1], c=agg_predictions, cmap='viridis')
plt.title("Agglomerative Clustering")

# Plot DBSCAN clustering
plt.subplot(133)
plt.scatter(X[:, 0], X[:, 1], c=dbscan_predictions, cmap='viridis')
plt.title("DBSCAN Clustering")

plt.show()
```
